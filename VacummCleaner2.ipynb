```python
class Environment:
    def __init__(self):
        # Initial state of rooms (Dirty or Clean)
        self.rooms = {
            "A": "Dirty",
            "B": "Dirty"
        }

    def get_status(self, location):
        return self.rooms[location]

    def clean_room(self, location):
        self.rooms[location] = "Clean"


class ModelBasedVacuumAgent:
    def __init__(self, environment):
        self.env = environment
        self.location = "A"  # Initial position
        self.model = {"A": None, "B": None}  # Internal memory

    def perceive(self):
        status = self.env.get_status(self.location)
        self.model[self.location] = status
        return status

    def act(self):
        status = self.perceive()

        # If current room is dirty, clean it
        if status == "Dirty":
            print(f"Cleaning Room {self.location}")
            self.env.clean_room(self.location)
            self.model[self.location] = "Clean"

        # If both rooms are clean, stop
        if self.model["A"] == "Clean" and self.model["B"] == "Clean":
            print("All rooms are clean. Stopping.")
            return False

        # Move to other room if needed
        if self.location == "A":
            self.location = "B"
        else:
            self.location = "A"

        print(f"Moving to Room {self.location}")
        return True


# Running the agent
env = Environment()
agent = ModelBasedVacuumAgent(env)

while agent.act():
    pass
```
