```python
def get_support_structure(block, state):
    """
    Returns the full support chain of a block.
    Example: A on B, B on C â†’ returns ['A','B','C']
    """
    support = []
    current = block

    while current != 'Table':
        support.append(current)
        current = state.get(current, 'Table')

    return support


def calculate_heuristic(state, goal):
    """
    Calculates global heuristic e(p)
    """
    heuristic = 0

    for block in state.keys():
        current_support = get_support_structure(block, state)
        goal_support = get_support_structure(block, goal)

        if current_support == goal_support:

            heuristic += len(current_support)
        else:

            heuristic -= len(current_support)

    return heuristic


if __name__ == "__main__":

    # Current State
    state = {
        'A': 'B',
        'B': 'Table',
        'C': 'A'
    }

    # Goal State
    goal = {
        'A': 'B',
        'B': 'C',
        'C': 'Table'
    }

    h_value = calculate_heuristic(state, goal)

    print("Heuristic Value e(p):", h_value)
```
