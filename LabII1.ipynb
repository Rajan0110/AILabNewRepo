```python
import random

class VacuumEnvironment:
    """Simulates a simple two-room world (A and B)."""
    def __init__(self):
        # Initial state: rooms are randomly Dirty (1) or Clean (0)
        self.rooms = {'A': random.randint(0, 1), 'B': random.randint(0, 1)}
        self.agent_location = random.choice(['A', 'B'])

    def get_percept(self):
        """Returns (current_location, status_of_location)."""
        status = 'Dirty' if self.rooms[self.agent_location] == 1 else 'Clean'
        return (self.agent_location, status)

    def execute_action(self, action):
        """Updates the environment based on the agent's action."""
        if action == 'Suck':
            self.rooms[self.agent_location] = 0  # Clean the room
        elif action == 'MoveRight':
            self.agent_location = 'B'
        elif action == 'MoveLeft':
            self.agent_location = 'A'

def reflex_agent(percept):
    """
    The Simple Reflex Agent Logic.
    Rules:
    1. If current room is Dirty -> Suck.
    2. If in Room A and Clean  -> Move Right.
    3. If in Room B and Clean  -> Move Left.
    """
    location, status = percept

    if status == 'Dirty':
        return 'Suck'
    elif location == 'A':
        return 'MoveRight'
    elif location == 'B':
        return 'MoveLeft'

# --- Simulation ---
env = VacuumEnvironment()
print(f"Initial State: {env.rooms} | Agent starting at: {env.agent_location}")

for step in range(1, 6):
    current_percept = env.get_percept()
    action = reflex_agent(current_percept)

    print(f"Step {step}: Location {current_percept[0]} is {current_percept[1]}. Agent chooses: {action}")

    env.execute_action(action)

print(f"Final State: {env.rooms}")
```
