```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.1, epochs=10):
        self.weight = 0
        self.bias = 0
        self.lr = learning_rate
        self.epochs = epochs

    def activation(self, x):
        return 1 if x > 0 else 0

    def predict(self, x):
        return self.activation(self.weight * x + self.bias)

    def train(self, X, y):
        for epoch in range(self.epochs):
            for i in range(len(X)):
                y_pred = self.predict(X[i])
                error = y[i] - y_pred

                self.weight += self.lr * error * X[i]
                self.bias += self.lr * error
            print(f"Epoch {epoch+1}: weight = {self.weight:.4f}, bias = {self.bias:.4f}")


if __name__ == "__main__":

    X = np.array([0, 1])
    y = np.array([1, 0])

    perceptron = Perceptron(learning_rate=0.1, epochs=10)
    perceptron.train(X, y)


    print("\nTesting NOT Gate Perceptron:")
    for x_input in X:
        print(f"Input: {x_input} -> Output: {perceptron.predict(x_input)}")
```
