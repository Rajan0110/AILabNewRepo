```python
import math

def acceptance_probability(delta_E, T):
    """
    Calculate probability of accepting inferior node
    """
    if T == 0:
        return 0
    return math.exp(-delta_E / T)


def temperature_schedule(initial_T, cooling_rate, iteration):
    """
    Exponential cooling schedule:
    T = T0 * (cooling_rate ^ iteration)
    """
    return initial_T * (cooling_rate ** iteration)


if __name__ == "__main__":

    delta_E = 5
    initial_T = 100
    cooling_rate = 0.85
    iterations = 20

    print("Iteration | Temperature | Acceptance Probability")
    print("-----------------------------------------------")

    for i in range(iterations):
        T = temperature_schedule(initial_T, cooling_rate, i)
        prob = acceptance_probability(delta_E, T)
        print(f"{i:9d} | {T:11.4f} | {prob:.6f}")
```
